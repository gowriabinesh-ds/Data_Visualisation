## Univariate Analysis

```python
import seaborn as sb
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

## Bivariate Analysis : Distribution Plot

### Joint plot

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset)
plt.show()
```

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='kde') # kde--> kernel density estimation
plt.show()
```

### Pair plot

```python
sb.pairplot(dataset,hue = 'gender',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

## For Categorical Plot

```python
sb.stripplot(x = "ssc_b", y = "ssc_p", data = dataset)
plt.show()
```

```python
sb.stripplot(x = "gender", y = "ssc_p", data = dataset,jitter=True)
plt.show()
```

```python
sb.swarmplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
```

<h4 style="color:darkblue;"> Swarm plot is used to show the distribution of individual data points across categories without overlapping.</h4>
<h4 style="color:darkblue;"> It helps clearly visualize how values are spread within each group and makes it easy to compare patterns or differences between categories.</h4>

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
sb.regplot(x = "total_bill", y = "tip", data = df)
sb.lmplot(x = "total_bill", y = "tip", data = df)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('anscombe')
sb.lmplot(x = "x", y = "y", data = df.query("dataset == 'II'"))
plt.show()
```
